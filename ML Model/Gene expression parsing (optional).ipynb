```python
import pandas as pd
import gzip

file_path = "/mnt/c/Users/Surface/Downloads/GSE2034_series_matrix.txt.gz"

with gzip.open(file_path, 'rt') as f:
    lines = f.readlines()

print("Total lines:", len(lines))

for i, line in enumerate(lines):
    if "ID_REF" in line:
        print("Matrix starts at line:", i)
        print("Header line:", line)
        break
    
file_path = "/mnt/c/Users/Surface/Downloads/GSE2034_series_matrix.txt.gz"
expression_df = pd.read_csv(
    file_path,
    compression='gzip',
    sep="\t",
    skiprows=54
)

print(expression_df.shape)
expression_df.head()

expression_df = expression_df.set_index("ID_REF")
print(expression_df.shape)
expression_df.head()

phenotype_lines = [line for line in lines if "relapse" in line.lower()]
for line in phenotype_lines:
    print(line)
    

for line in lines:
    if line.startswith("!Sample_characteristics_ch1"):
        if "bone relapses" in line:
            phenotype_line = line
            break


phenotype_values = phenotype_line.strip().split("\t")[1:]


labels = [int(x.split(":")[1].strip().replace('"','')) for x in phenotype_values]

print("Number of labels:", len(labels))
print("First 10 labels:", labels[:10])


X = expression_df.T
y = labels


gene_variances = X.var(axis=0)
top_genes = gene_variances.sort_values(ascending=False).head(1000).index

X_reduced = X[top_genes]

print(X_reduced.shape)

X_reduced.to_csv("/mnt/c/Users/Surface/Downloads/GSE2034_reduced_expression.csv")
pd.Series(y).to_csv("/mnt/c/Users/Surface/Downloads/GSE2034_labels.csv")
```
